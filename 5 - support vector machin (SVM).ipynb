```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
clf = SVC(kernel = 'linear')
```

```python
URL_='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
data_iris = pd.read_csv(URL_,header = None)

X = data_iris.iloc[:,:2]
Y = data_iris[4]
```

```python
features_ = np.asarray(X)
labels_ = np.asarray(Y)
```

```python
svm = clf.fit(features_,labels_)
```

```python
test_ = [[5.1, 3.5]]
clf.predict(test_)
```
